
# Q4 : 간단 계산기


```python
import math


def read_int(prompt: str) -> int:
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("오류: 정수가 아닙니다. 정수만 입력할 수 있습니다.")
            print("잘못된 입력입니다. 정수를 입력해주세요.")


def calculate(a: int, b: int, op: str):
    if op == "+":
        return a + b
    if op == "-":
        return a - b
    if op == "*":
        return a * b
    if op == "/":
        if b == 0:
            raise ZeroDivisionError("0으로 나눌 수 없습니다.")
        return a / b
    if op == "**":
        return math.pow(a, b)  # math 모듈로 제곱 연산 지원

    raise ValueError(f"지원하지 않는 연산자입니다: {op}")


def main():
    print("간단 계산기 (+, -, *, /, **)")

    while True:
        a = read_int("첫 번째 정수를 입력하세요: ")
        b = read_int("두 번째 정수를 입력하세요: ")
        op = input("연산자를 입력하세요 (+, -, *, /, **): ").strip()

        try:
            result = calculate(a, b, op)
            print("결과:", result)
        except ZeroDivisionError as e:
            print("오류:", e)
        except ValueError as e:
            print("오류:", e)

        again = input("계속 계산할까요? (y/n): ").strip().lower()
        if again != "y":
            print("계산기를 종료합니다.")
            break


if __name__ == "__main__":
    main()
```
