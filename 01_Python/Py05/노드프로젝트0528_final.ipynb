# 키오스크 만들기(샌드위치 판매점)

```python
import datetime
 
 
class KIOSKI:
    def __init__(self):
        self.item = {"name": "샌드위치", "price": 4500}
        self.qty = 0
 
    def _read_int(self, prompt: str, min_value: int | None = None, max_value: int | None = None) -> int:
        while True:
            raw = input(prompt).strip()
            try:
                value = int(raw)
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해 주세요.")
                continue
 
            if min_value is not None and value < min_value:
                print(f"{min_value} 이상을 입력해 주세요.")
                continue
            if max_value is not None and value > max_value:
                print(f"{max_value} 이하를 입력해 주세요.")
                continue
 
            return value
 
    def order(self) -> None:
        print("\n[메뉴판]")
        print(f"1. {self.item['name']} - {self.item['price']}원")
        print("0. 뒤로가기")
 
        choice = self._read_int("메뉴 번호를 선택하세요: ", 0, 1)
        if choice == 0:
            return
 
        self.qty = self._read_int("수량을 입력하세요: ", 1, 99)
        print(f"주문 완료: {self.item['name']} x{self.qty}")
 
    def get_total_amount(self) -> int:
        return self.item["price"] * self.qty
 
    def print_order_summary(self) -> None:
        if self.qty <= 0:
            print("\n주문 내역이 없습니다.")
            return
 
        print("\n[주문 메뉴]")
        line_total = self.get_total_amount()
        print(f"1. {self.item['name']} x{self.qty} = {line_total}원")
        print(f"총 주문금액: {line_total}원")
 
    def _read_card_number(self) -> str:
        while True:
            card_no = input("카드번호 16자리를 입력하세요(숫자만): ").strip().replace(" ", "")
            if len(card_no) != 16 or not card_no.isdigit():
                print("잘못된 카드번호입니다. 16자리 숫자를 입력해 주세요.")
                continue
            return card_no
 
    def pay(self) -> tuple[str, int]:
        if self.qty <= 0:
            raise ValueError("주문 내역이 없어 결제할 수 없습니다.")
 
        total = self.get_total_amount()
        print("\n[지불]")
        print(f"결제 금액: {total}원")
        print("1. 카드")
        print("2. 현금")
 
        method = self._read_int("결제 수단을 선택하세요: ", 1, 2)
        if method == 1:
            print("\n[카드 결제]")
            print("1. 카드번호 입력")
            print("2. 카드 삽입(Enter로 진행)")
            card_read_method = self._read_int("카드 인식 방법을 선택하세요: ", 1, 2)
            if card_read_method == 1:
                _ = self._read_card_number()
                print("카드번호가 확인되었습니다.")
            else:
                input("카드를 넣어 주세요. 준비되면 Enter를 누르세요: ")
                print("카드가 인식되었습니다.")
 
            print("카드 결제가 완료되었습니다.")
            return "카드", 0
 
        while True:
            cash = self._read_int("현금을 넣어주세요: ", 0)
            if cash < total:
                print(f"금액이 부족합니다. {total - cash}원 더 필요합니다.")
                continue
            change = cash - total
            print(f"현금 결제가 완료되었습니다. 거스름돈: {change}원")
            return "현금", change
 
    def print_receipt(self, payment_method: str, change: int) -> None:
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print("\n========== 주문표 ==========")
        print(f"주문시간: {now}")
        print("----------------------------")
        line_total = self.get_total_amount()
        print(f"{self.item['name']} x{self.qty}  {line_total}원")
        print("----------------------------")
        total = self.get_total_amount()
        print(f"주문금액: {total}원")
        print(f"결제수단: {payment_method}")
        if payment_method == "현금":
            print(f"거스름돈: {change}원")
        print("============================\n")
 
    def reset(self) -> None:
        self.qty = 0
 
    def run(self) -> None:
        print("샌드위치 키오스크에 오신 것을 환영합니다.")
        while True:
            print("\n[메인]")
            print("1. 주문하기")
            print("2. 주문내역/금액 확인")
            print("3. 지불")
            print("0. 종료")
 
            choice = self._read_int("번호를 선택하세요: ", 0, 3)
 
            if choice == 0:
                print("프로그램을 종료합니다.")
                return
 
            if choice == 1:
                self.reset()
                self.order()
                if self.qty > 0:
                    self.print_order_summary()
                    try:
                        method, change = self.pay()
                        self.print_receipt(method, change)
                        self.reset()
                    except ValueError as e:
                        print("오류:", e)
            elif choice == 2:
                self.print_order_summary()
            elif choice == 3:
                try:
                    method, change = self.pay()
                    self.print_receipt(method, change)
                    self.reset()
                except ValueError as e:
                    print("오류:", e)
 
 
KIOSKI().run()
```
